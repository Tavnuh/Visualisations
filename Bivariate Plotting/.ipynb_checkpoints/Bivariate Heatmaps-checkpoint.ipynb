```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

fuel_econ = pd.read_csv('fuel-econ.csv')
```

### Matplotlib - Hist2d

```python
plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb',bins=20,cmin=0.5,
          cmap= 'viridis_r')
plt.colorbar()
plt.xlabel('Displacement')
plt.ylabel('Combined Fuel efficiency');
```
