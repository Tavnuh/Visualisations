```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
pokemon = pd.read_csv('pokemon.csv')
```

It can be useful to create new variables as functions of existing variables. Like the ratio of one to another, the log or exponent of a variable etc

```python
pokemon['atk_ratio'] = pokemon['attack'] / pokemon['special-attack']
pokemon['def_ratio'] = pokemon['defense'] / pokemon['special-defense']
```

```python
plt.scatter(data = pokemon, x = 'atk_ratio', y = 'def_ratio',alpha = 1/3)
plt.xlabel('Offensive Bias (Physical vs Special)')
plt.ylabel('Defensive Bias (Physical vs Special)')
```

Now we'll make the scale log on both axes, and add tick marks appropriate for the scale

```python
plt.scatter(data = pokemon, x = 'atk_ratio', y = 'def_ratio',alpha = 1/3)
plt.xlabel('Offensive Bias (Physical vs Special)')
plt.ylabel('Defensive Bias (Physical vs Special)')
plt.xscale('log')
plt.yscale('log')

tick_loc = [0.25,0.5,1,2,4]
plt.xticks(tick_loc,tick_loc)
plt.yticks(tick_loc,tick_loc);
```

And set the axis limits to remove extreme outliers

```python
plt.scatter(data = pokemon, x = 'atk_ratio', y = 'def_ratio',alpha = 1/3)
plt.xlabel('Offensive Bias (Physical vs Special)')
plt.ylabel('Defensive Bias (Physical vs Special)')
plt.xscale('log')
plt.yscale('log')

tick_loc = [0.25,0.5,1,2,4]
plt.xticks(tick_loc,tick_loc)
plt.yticks(tick_loc,tick_loc)
plt.xlim(2 ** -2.5, 2 ** 2.5)
plt.ylim(2 ** -2.5, 2 ** 2.5);
```
