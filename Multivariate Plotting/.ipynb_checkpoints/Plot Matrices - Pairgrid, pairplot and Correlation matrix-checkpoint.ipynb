```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
pokemon = pd.read_csv('pokemon.csv')
```

### Seaborn pairgrid with scatter plots mapped

```python
pkmn_stats = ['hp','attack','defense','special-attack','special-defense']
g = sns.PairGrid(data = pokemon, vars = pkmn_stats)
g = g.map_offdiag(plt.scatter)
g.map_diag(plt.hist)
```

Or, an easier way (use kind and diag_kind to change the plot types on the off_diag and diag positions)

```python
sns.pairplot(data = pokemon[pkmn_stats]);
```

```python
pokemon[pkmn_stats].corr()
```

```python
sns.heatmap(pokemon[pkmn_stats].corr(), cmap = 'rocket_r',annot = True,
           vmin = 0)
```
